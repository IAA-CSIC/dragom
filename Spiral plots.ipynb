```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import profiles as prof
import spirals as sp
import parameters as par
from astropy import units as u
from scipy.optimize import curve_fit
```

```python
directory = '../SVS13/espirales/'
```

# S1

```python
#import data
file_nameA1 = 'DistA-S1.dat'
file_nameB1 = 'DistB-S1.dat'

dataA1 = np.loadtxt(directory+file_nameA1)
angA1 = [d[0] for d in dataA1]
distA1 = [d[1] for d in dataA1]
errA1 = [d[2] for d in dataA1]
nangA1 = prof.new_angle(angA1,180)

dataB1 = np.loadtxt(directory+file_nameB1)
angB1 = [d[0] for d in dataB1]
distB1 = [d[1] for d in dataB1]
errB1 = [d[2] for d in dataB1]
nangB1 = prof.new_angle(angB1,180)

# lets fit
ifitA1 = 7
jfitA1 = len(distA1)
lineA1, resA1, _, _, _ = np.polyfit(distA1[ifitA1:jfitA1], nangA1[ifitA1:jfitA1], 1, full=True)

ifitB1 = 0
jfitB1 = len(distB1)
lineB1, resB1, _, _, _ = np.polyfit(distB1[ifitB1:jfitB1], nangB1[ifitB1:jfitB1], 1, full=True)

r = np.linspace(0.15,0.6,100)
ang_fitA1 = lineA1[0]*r+lineA1[1]
ang_fitB1 = lineB1[0]*r+lineB1[1]


#lets plot
plt.figure(figsize=(20,10))
plt.plot(r,ang_fitA1,'--b',label=f'Linear fit: m={round(lineA1[0])}, n={round(lineA1[1])}')
plt.plot(r,ang_fitB1,'--r',label=f'Linear fit: m={round(lineB1[0])}, n={round(lineB1[1])}')
plt.errorbar(distA1,nangA1, xerr=0,fmt='ob',capsize=3,label='S1 (dist to A)')
plt.errorbar(distB1,nangB1, xerr=0,fmt='or',capsize=3,label='S1 (dist to B)')
#plt.xlim(-0.01,0.6)
#plt.ylim(-5,5)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to star (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
plt.legend(fontsize=18)
plt.show()
```

# S2-distA

```python
#import data
file_nameA2 = 'DistA-S2.dat'
file_nameA2_d = 'DistA-S2-duda.dat'

dataA2 = np.loadtxt(directory+file_nameA2)
angA2 = [d[0] for d in dataA2]
distA2 = [d[1] for d in dataA2]
errA2 = [d[2] for d in dataA2]

dataA2_d = np.loadtxt(directory+file_nameA2_d)
angA2_d = [d[0] for d in dataA2_d]
distA2_d = [d[1] for d in dataA2_d]
errA2_d = [d[2] for d in dataA2_d]
nangA2_d = prof.new_angle(angA2_d,180)

# lets fit
ifitA2 = 0
jfitA2 = len(distA2)-6
lineA2, resA2, _, _, _ = np.polyfit(distA2[ifitA2:jfitA2], angA2[ifitA2:jfitA2], 1, full=True)

ifitA2_d = 0
jfitA2_d = len(distA2_d)
lineA2_d, resA2_d, _, _, _ = np.polyfit(distA2_d[ifitA2_d:jfitA2_d], nangA2_d[ifitA2_d:jfitA2_d], 1, full=True)

r = np.linspace(0.68,0.88,100)
r_d = np.linspace(0.75,1.4,100)
angA2_fit = lineA2[0]*r+lineA2[1]
angA2_fit_d = lineA2_d[0]*r_d+lineA2_d[1]


#lets plot
plt.figure(figsize=(20,10))
plt.plot(r,angA2_fit,'--b',label=f'Linear fit: m={round(lineA2[0])}, n={round(lineA2[1])}')
plt.plot(r_d,angA2_fit_d,'--r',label=f'Linear fit: m={round(lineA2_d[0])}, n={round(lineA2_d[1])}')
plt.errorbar(distA2, angA2, xerr=0,fmt='ob',capsize=3,label='S2')
plt.errorbar(distA2_d, nangA2_d, xerr=0,fmt='or',capsize=3,label='S2-duda')
#plt.xlim(-0.01,0.6)
#plt.ylim(-5,5)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
plt.legend(fontsize=18)
plt.show()
```

# S2-distB

```python
#import data
file_nameB2 = 'DistB-S2.dat'
file_nameB2_d = 'DistB-S2-duda.dat'

dataB2 = np.loadtxt(directory+file_nameB2)
angB2 = [d[0] for d in dataB2]
distB2 = [d[1] for d in dataB2]
errB2 = [d[2] for d in dataB2]

dataB2_d = np.loadtxt(directory+file_nameB2_d)
angB2_d = [d[0] for d in dataB2_d]
distB2_d = [d[1] for d in dataB2_d]
errB2_d = [d[2] for d in dataB2_d]
nangB2_d = prof.new_angle(angB2_d,180)

# lets fit linear
ifitB2 = 0
jfitB2 = len(distB2)
lineB2, resB2, _, _, _ = np.polyfit(distB2[ifitB2:jfitB2], angB2[ifitB2:jfitB2], 1, full=True)

ifitB2_d = 0
jfitB2_d = len(distB2_d)
lineB2_d, resB2_d, _, _, _ = np.polyfit(distB2_d[ifitB2_d:jfitB2_d], nangB2_d[ifitB2_d:jfitB2_d], 1, full=True)

r = np.linspace(0.45,0.65,100)
r_d = np.linspace(0.75,1.6,100)
angB2_fit = lineB2[0]*r+lineB2[1]
angB2_fit_d = lineB2_d[0]*r_d+lineB2_d[1]

# lets fit exponential
popt_exp, pcov_exp = curve_fit(prof.spiral_exp, distB2+distB2_d, angB2+nangB2_d)
r_exp = np.linspace(0.45,1.6,100)
angB2_fit_exp = prof.spiral_exp(r_exp, popt_exp[0],popt_exp[1],popt_exp[2])


#lets plot
plt.figure(figsize=(20,10))
plt.plot(r,angB2_fit,'--b',label=f'Linear fit: m={round(lineB2[0])}, n={round(lineB2[1])}')
plt.plot(r_d,angB2_fit_d,'--r',label=f'Linear fit: m={round(lineB2_d[0])}, n={round(lineB2_d[1])}')
plt.plot(r_exp, angB2_fit_exp,'--',label=f'Exponential fit: a={round(popt_exp[0])}, b={round(popt_exp[1])},c={round(popt_exp[2])}')
plt.errorbar(distB2, angB2, xerr=0,fmt='ob',capsize=3,label='S2')
plt.errorbar(distB2_d, nangB2_d, xerr=0,fmt='or',capsize=3,label='S2-duda')
#plt.xlim(-0.01,0.6)
#plt.ylim(-5,5)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to B (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
plt.legend(fontsize=18)
plt.show()
```

# S3-distA

```python
file_nameA3 = 'DistA-S3.dat' 
file_nameA3_d = 'DistA-S3-duda.dat' 

dataA3 = np.loadtxt(directory+file_nameA3)[4:]
angA3 = [d[0] for d in dataA3]
distA3 = [d[1] for d in dataA3]

dataA3_d = np.loadtxt(directory+file_nameA3_d)
angA3_d = [d[0] for d in dataA3_d]
nangA3_d = prof.new_angle(angA3_d,180)
distA3_d = [d[1] for d in dataA3_d]

r = np.arange(0.001,1.5,0.001) #arcsec

#let's fit linear
ifitA3 = 0
jfitA3 = len(distA3)-9
lineA3, resA3, _, _, _ = np.polyfit(distA3[ifitA3:jfitA3], angA3[ifitA3:jfitA3], 1, full=True)
angA3_fit = lineA3[0]*r+lineA3[1]


plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.plot(distA3,angA3,'ob',label='S3')
plt.plot(distA3_d, nangA3_d,'or', label='S3-duda')
plt.plot(r,angA3_fit,'--b',label=f'Linear fit: m={round(lineA3[0])}, n={round(lineA3[1])}')
#plt.xlim(-0.01,1.5)
#plt.ylim(-5,200)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=18)
plt.legend(fontsize=18)
plt.show()




# lets fit the spiral
f = np.pi/180
popt, pcov = curve_fit(sp.spiral_M02, distA3, angA3, bounds=([0,0,0],[100,360*f,1]))


plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.plot(distA3,angA3,'ob',label='S3')
plt.plot(distA3_d, nangA3_d,'or', label='S3-duda')
P_us = [sp.spiral_M02(r0,popt[0],popt[1],popt[2]) for r0 in r] #radianes
plt.plot(r, P_us, '--',label=f'rc={round(popt[0],4)} arcsec, Pc={round(popt[2]/f)} deg, hc={round(popt[1],4)}')
#plt.xlim(-0.01,1.5)
#plt.ylim(-5,200)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=18)
plt.legend(fontsize=18)
plt.show()

######################################
rc0=popt[0]
hc0=popt[1]
Pc0=popt[2]


rc_us = np.arange(0.5,15,3) #arcsec
Pc_us = np.arange(0,360,90)*f #radianes
hc_us = np.arange(0.005,0.025,0.005) #arcsec
#hc_us = (par.H_scale(rc_us*300*u.au, M_star = 0.41*u.M_sun, T_disk = 126*u.K)/(rc_us*300*u.au)).decompose().value

######################################
#dependence with rc
plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.plot(distA3,angA3,'ob',label='S3')
plt.plot(distA3_d, nangA3_d,'or', label='S3-duda')

for _, d in enumerate(rc_us):
    spiral = [sp.spiral_M02(r0,d,hc0,Pc0) for r0 in r] #radianes
    plt.plot(r, spiral, '--',label=f'rc={round(d,4)} arcsec, Pc={round(Pc0/f)} deg, hc={round(hc0,4)}')

#plt.xlim(-0.01,1.5)
#plt.ylim(-5,200)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=18)
plt.legend(fontsize=18)
plt.show()

#dependence with Pc
plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.plot(distA3,angA3,'ob',label='S3')
plt.plot(distA3_d, nangA3_d,'or', label='S3-duda')

for _, a in enumerate(Pc_us):
    spiral = [sp.spiral_M02(r0,rc0,hc0,a) for r0 in r] #radianes
    plt.plot(r, spiral, '--',label=f'rc={round(rc0,4)} arcsec, Pc={round(a/f)} deg, hc={round(hc0,4)}')

#plt.xlim(-0.01,1.5)
#plt.ylim(-5,200)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=18)
plt.legend(fontsize=18)
plt.show()

#dependence with hc
plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.plot(distA3,angA3,'ob',label='S3')
plt.plot(distA3_d, nangA3_d,'or', label='S3-duda')

for _, h in enumerate(hc_us):
    spiral = [sp.spiral_M02(r0,rc0,h,Pc0) for r0 in r] #radianes
    plt.plot(r, spiral, '--',label=f'rc={round(rc0,4)} arcsec, Pc={round(Pc0/f)} deg, hc={round(h,4)}')

#plt.xlim(-0.01,1.5)
#plt.ylim(-5,200)
rc('xtick', labelsize=20) 
rc('ytick', labelsize=20) 
plt.xlabel('Distance to A (arcsec)',fontsize=20)
plt.ylabel('PA (deg)',fontsize=20)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=18)
plt.legend(fontsize=18)
plt.show()
```
