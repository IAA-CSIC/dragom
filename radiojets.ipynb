```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_all='jets_Guillem.dat'
file_Tob='jets_OMC2.dat'
file_mine='jets_OMC2_us.dat'
#file_mine='jets-OMC-2,dat'
with open(file_all) as f:
    file_all = f.readlines()[1:]
with open(file_Tob) as f:
    file_Tob_us = f.readlines()[1:]
with open(file_mine) as f:
    file_us = f.readlines()[1:]
```

```python
d=0.386
dist_all = [float(f.split(' ')[1]) for f in file_all]
Lbol_all = [float(f.split(' ')[2]) for f in file_all]
Lbol_Tob_us = [float(f.split(' ')[2]) for f in file_Tob_us]
dist_Tob_us = np.full(len(Lbol_Tob_us),d)
Lbol_us = [float(f.split(' ')[2]) for f in file_us]
dist_us = np.full(len(Lbol_us),d)
```

```python
plt.figure()
plt.plot(dist_all,Lbol_all,'ok')
#plt.plot(dist_Tob_us,Lbol_Tob_us,'ob')
plt.plot(dist_us,Lbol_us,'or')
plt.ylim(-2,50)
plt.xlim(0.1,1)
plt.xlabel('distance (kpc)')
plt.ylabel('Lbol (Lsol)')
plt.show()
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(dist_all,Lbol_all,'ok')
ax.plot(dist_Tob_us,Lbol_Tob_us,'ob')
ax.plot(dist_us,Lbol_us,'or')
ax.set_yscale('log')
ax.set_ylim(0.01,10)
ax.set_xlim(0.1,0.4)
ax.set_xlabel('distance (kpc)')
ax.set_ylabel('Lbol (Lsol)')
```
